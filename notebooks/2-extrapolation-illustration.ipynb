```python
import pandas as pd 
import numpy as np
from tqdm import tqdm
from matplotlib import pyplot as plt 
import seaborn as sns
sns.set()
```

```python
np.random.seed(42)
```

```python
n_samples = 500
mu = [1, 1]
r = [[1, 0.9],
    [0.9, 1]
    ]
x = np.random.multivariate_normal(mu, r, size=n_samples)
```

```python
pd.DataFrame(x).corr()
```

```python
plt.figure(figsize=(12, 8))
plt.plot(x[:, 0], x[:, 1], ".b")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

```python
x_shuffled = x.copy()
np.random.shuffle(x_shuffled[:, 1])
```

```python
plt.figure(figsize=(12, 8))
plt.plot(x[:, 0], x[:, 1], ".b", label="No shuffle")
plt.plot(x_shuffled[:, 0], x_shuffled[:, 1], ".r", label="Shuffled x1", alpha=0.5)
plt.xlabel("x1 and x1 shuffled")
plt.ylabel("x2")
plt.legend()
plt.show()
```
